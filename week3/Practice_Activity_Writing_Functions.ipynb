```python
def plot_species_island(species = None, island=None):
    """
    Dynamic lookup
    """
    new_penguins = penguins

    if species is not None:
        new_penguins = new_penguins[new_penguins['species'] == species] #Filter for species 
    
    if island is not None:
        new_penguins = new_penguins[new_penguins["island" == island]] #Filter for island 

    plot = (
        ggplot(new_penguins, aes(x= "bill_length_mm", y= "bill_depth_mm", fill= "island"))
        + geom_point()
        + labs(
            title = "Scatter Plot",
            x= "bill_length", y= "bill_depth"
        )
    )
    
    plot.show()
```

```python
plot_species_island(species= "Adelie")
```

```python
def rescale_vector(vec):
    if any(np.isnan(vec)):
        exit("Stop now")

    minimum_vec = min(vec)
    print("minimum: " + str(minimum_vec))

    maximum_vec = max(vec)
    print("maximum: " + str(maximum_vec))

    rescale = (vec - minimum_vec) / (maximum_vec - minimum_vec)

    return rescale
```

```python
#trials from lecture 
my_list = [0, 1, 2, 3, 4]
new_list = np.sqrt(my_list)
print(rescale_vector(new_list))
```

```python
my_list = [-1, 0, 1,2,3,4]
new_list = np.sqrt(my_list)
print(rescale_vector(new_list))
```
